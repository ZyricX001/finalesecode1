```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
# ===================== DATA LOADING =====================
train = pd.read_csv('/kaggle/input/mse-2-ai-201-b-aiml-a/sample_submission.csv')
test = pd.read_csv('/kaggle/input/mse-2-ai-201-b-aiml-a/test.csv')

print(train.head())
print(train.info())
print(train.isnull().sum())
print(train['Class'].value_counts())
```

```python
# ===================== TARGET & FEATURES =====================
y = train["Class"]
x = train.drop("Class", axis=1)

num_cols = x.select_dtypes(include=["int64", "float64"]).columns
cat_cols = x.select_dtypes(include=["object"]).columns
```

```python
 ===================== DATA VISUALIZATION =====================
plt.figure(figsize=(10,4))
train.isnull().sum().plot(kind="bar")
plt.title("Missing Values")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x=y)
plt.title("Target Distribution")
plt.show()
```

```python
# ===================== CORRELATION ANALYSIS (ADDED AT RIGHT PLACE) =====================
plt.figure(figsize=(10,6))
sns.heatmap(train.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# ===================== DATA CLEANING =====================
for col in num_cols:
    x[col] = x[col].fillna(x[col].median())
    test[col] = test[col].fillna(x[col].median())

for col in cat_cols:
    if x[col].mode().empty:
        fill_value = "Missing"
    else:
        fill_value = x[col].mode()[0]
    x[col] = x[col].fillna(fill_value)
    test[col] = test[col].fillna(fill_value)
```

```python
# ===================== OUTLIER ANALYSIS =====================
x[num_cols].boxplot(figsize=(12,4))
plt.title("Outlier Boxplot")
plt.show()
```

```python
# ===================== ENCODING TARGET =====================
le = LabelEncoder()
y_enc = le.fit_transform(y)
print("Classes:", le.classes_)
```

```python
# ===================== PREPROCESSING PIPELINE =====================
pre = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ("num", "passthrough", num_cols)
])

model = Pipeline([
    ("prep", pre),
    ("rf", RandomForestClassifier(random_state=42))
])
```

```python
# ===================== TRAIN TEST SPLIT =====================
x_train, x_val, y_train, y_val = train_test_split(
    x, y_enc, test_size=0.2, random_state=42
)
```

```python
# ===================== MODEL TRAINING =====================
model.fit(x_train, y_train)
```

```python
# ===================== HYPERPARAMETER TUNING =====================
params = {
    "rf__n_estimators": [100, 200],
    "rf__max_depth": [None, 10, 20]
}

tune = RandomizedSearchCV(
    model,
    params,
    n_iter=4,
    cv=3,
    scoring="f1_macro",
    n_jobs=-1
)

tune.fit(x, y_enc)
model = tune.best_estimator_
```

```python
# ===================== MODEL EVALUATION (ADDED AT RIGHT PLACE) =====================
val_preds = model.predict(x_val)
acc = accuracy_score(y_val, val_preds)
print("Validation Accuracy:", acc)
```

```python
# ===================== FINAL TRAINING =====================
model.fit(x, y_enc)
```

```python
# ===================== TEST PREDICTION =====================(for single class)
preds = model.predict(test)
pred_labels = le.inverse_transform(preds)
```

```python
# ===================== SUBMISSION FILE =====================
submission = pd.DataFrame({
    "id": test["id"],
    "Class": pred_labels
})

submission.to_csv("abhilash1.csv", index=False)
print("\nSubmission file created successfully!")
submission.head()
```
